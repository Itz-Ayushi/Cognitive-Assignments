```python
import numpy as np
import pandas as pd

rno=102317237
np.random.seed(rno)
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
categories = ["Electronics", "Clothing", "Home & Kitchen", "Sports"]

sales_data=np.random.randint(1000, 5001, size=(12, 4))

df = pd.DataFrame(sales_data, columns=categories, index=months)

print(df.head())
print(df.describe())


totals = df.sum()
print(totals)
totalm = df.sum(axis=1)
print(totalm)
growth_rates = df.pct_change().mean() * 100


df["Total Sales"] = totalm
df["Growth Rate"] = df["Total Sales"].pct_change() * 100

print(df)

discounted_df = df.copy()
if rno % 2 == 0:
    discounted_df["Electronics"] *= 0.9
else:
    discounted_df["Clothing"] *= 0.85

discounted_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
df.plot(kind='line', marker='o', title='Monthly Sales Trends')
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend()
plt.show()

# Boxplot
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[categories])
plt.title("Sales Distribution by Category")
plt.show()
```

```python
array = np.array([[1, -2, 3],[-4, 5, -6]])
print("Element-wise absolute:", np.abs(array))

flattened_array = array.flatten()
print("Percentiles:", np.percentile(flattened_array, [25, 50, 75]))
print("Mean:", np.mean(flattened_array))
print("Median:", np.median(flattened_array))
print("Standard Deviation:", np.std(flattened_array))
```
