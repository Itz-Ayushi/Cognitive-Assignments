```python
import numpy as np
avg_salary = np.mean(empdata['Salary'])
bonus = np.sum(empdata['Bonus'])
emp_age = np.min(empdata['Age'])
rate = np.max(empdata['Rating'])
print(avg_salary)
print(bonus)
print(emp_age)
print(rate)
```

```python
empdata.sort_values(by = ['Salary'],ascending = False)
```

```python
def categorize_performance(rating):
  if rating >= 4.5:
    return 'Excellent'
  elif rating >= 4.0:
    return 'Good'
  else:
    return 'Average'

empdata['Performance Category'] = empdata['Rating'].apply(categorize_performance)
empdata
```

```python
def deduct_tax(salary):
  return salary*0.1

empdata['Tax'] = empdata['Salary'].apply(deduct_tax)
empdata
```
