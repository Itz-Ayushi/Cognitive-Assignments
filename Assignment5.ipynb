```python
import numpy as np
gfg = np.matrix('[4, 1, 9; 12, 3, 1; 4, 5, 6]')
print('Sum of all elements:',np.sum(gfg))
print('Sum of all elements row-wise:',np.sum(gfg, axis=1))
print('Sum of all elements column-wise:',np.sum(gfg, axis=0))
```

```python
#a
array = np.array([10, 52, 62, 16, 16, 54, 453])
print('Sorted array:',np.sort(array))
print('Sorted array Indices:',np.argsort(array))
print('4 Smallest elements: ',np.sort(array)[:4])
print('5 Largest elements: ',np.sort(array)[-5:])

#b
array = np.array([1.0, 1.2, 2.2, 2.0, 3.0, 2.0])
print('Integer elements:',array[array%1==0])
print('Float elements:',array[array%1!=0])
```

Ques 3

```python
#a
name="AS"
X=0
for i in name:
    X=X+ord(i)
print("X:",X)
sales=np.array([X, X+50, X+100, X+150, X+200])
print(sales)

#b
tax_rate=((X % 5) + 5) / 100
tax=sales*tax_rate
print("tax:",tax)

#c
sales=sales+tax
discount=np.where(sales<(X+100),sales*0.95,sales*0.9)
print("Sales after discount:",discount)

#d
weeklySales=np.tile(discount,(3,1))
increase=np.array([[1],[1.02],[1.0404]])
print("Weekly Sales:", weeklySales*increase)
```

```python
import matplotlib.pyplot as plt
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
plt.title('Quadratic Function (y=x^2)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

y=np.sin(x)
plt.plot(x,y)
plt.title('Sin Function (y=sin(x))')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

y=np.exp(x)
plt.plot(x,y)
plt.title('Exponential Function (y=e^x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()

y=np.log(abs(x)+1)
plt.plot(x,y)
plt.title('Logarithmic Function (log|x|+1)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
